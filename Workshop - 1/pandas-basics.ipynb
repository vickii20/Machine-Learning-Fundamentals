1 - Total spend per customer

Find total amount spent by each customer (only those who have transactions). 
Return the top 10 customers by total spend.

5 - Average basket size and high-value customers

Compute for each customer:
	•	avg_basket_value = sum(amount)/count(transactions)
    
Then list customers with avg_basket_value > 2000 and at least 5 transactions, sorted by avg_basket_value desc.


```python
customer.head()
```

```python
transaction.head()
```

```python
# avg_basket_value = sum(amount) /  count(transactions)
# avg_basket_value > 2000 and at least 5 transactions, sorted by avg_basket_value desc.

transaction['avg_basket_value'] = (
    transaction.groupby(['customer_id'])['amount'].transform('sum') 
    / transaction.groupby(['customer_id'])['transaction_id'].transform('nunique')
)
```

```python
transaction.head()
```

```python
transaction['tnx_count'] = transaction.groupby(['customer_id'])['transaction_id'].transform('count')
transaction.head()
```

```python
sol = transaction.groupby(['customer_id','tnx_count','avg_basket_value']).agg(
    total_amount = ('amount' , 'sum')
).reset_index()

sol
```

```python
sol[ (sol['avg_basket_value'] > 2000) & (sol['tnx_count'] >= 5)].sort_values('avg_basket_value', ascending = False)
```

```python
data = data.dropna()
```

```python
data
```

```python
Join cust and txn (only matching customers) to compute per-city:

	•	total customers (distinct),

	•	total transactions,

	•	total revenue,

	•	average revenue per customer.


Return cities sorted by total revenue descending.
```

```python
a = data.groupby(['city']).agg(
    total_customers = ('customer_id','nunique'),
    total_transactions =('transaction_id' , 'count'),
    total_revenue = ('amount', 'sum'),
).reset_index()
a
```

9 - Detect likely duplicate customers (simple heuristic) and merge

Find potential duplicate customers in cust where first_name and last_name match but customer_id differs and signup_date within 30 days. Produce a list of duplicate pairs and suggest the survivor_customer_id (earliest signup).

```python
customer.head()
```

```python
transaction.head()
```

```python
cust = customer.copy()
cust
```

```python
val = pd.merge(customer,cust, left_on = ['first_name','last_name'], right_on = ['first_name','last_name'], how = 'inner')
val
```

```python
val = val[val['customer_id_x'] != val['customer_id_y']]
```

```python
val
```

```python
val.loc[:, 'sign_up_diff_days'] = (
    val['signup_date_x'] - val['signup_date_y']
).abs().dt.days
```

```python
val
```

```python
duplicates = val[val['sign_up_diff_days'] <= 30]
```

```python
duplicates
```

```python
duplicates[[
    'customer_id_x','customer_id_y',
    'first_name','last_name',
    'signup_date_x','signup_date_y',
    'sign_up_diff_days'
]]
```
