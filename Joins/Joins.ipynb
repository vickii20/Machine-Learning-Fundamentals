# Example for the joins

```python
data_tb1 = {'a1' : [1,2,3,4], 'b1' : ['a','b','c','d']}
data_tb2 = {'a2' : [2,3], 'b2' : ['b','c']}
```

```python
df1 = pd.DataFrame(data_tb1)
df2 = pd.DataFrame(data_tb2)
```

```python
df1
```

```python
df2
```

```python
# inner join concept : ("common values")

# this inner join were compare the a1-column to the a2-column, 
# if it is match with the a1-column it will print otherwise it will ignore

pd.merge(df1,df2, left_on = 'a1', right_on = 'a2', how = 'inner')

# this is like selecting the prefered column to compare it (left_on = 'a1', right_on = 'a2')
# how = 'inner' this is the condition to perform the operation.
```

```python
# left join concept : ("left Dominant side")

# this left join were compare the a1-column to the a2-column,
# if it is match with the a1-column it will print 
# also it does't match with the column it will print with null value (NAN Value)
# ** it will print the all rows whether the column were match or not **

pd.merge(df1,df2, left_on = 'a1', right_on = 'a2', how = 'left')
```

```python
# Right join concept : ("right Dominant side")

# this right join were compare the u_id to user_id, 
# if it is match with the u_id it will print
# also it does't match with the column it will print with null value (NAN Value)
# ** it will print the all rows whether the column were match or not **

tb1 = {'user_id': ['u1','u2','u3','u4'], 'name' : ['n1','n2','n3','n4']}
tb2 = {'u_id' : ['u2','u3','u4','u10'], 'age' : [20,30,40,100]}
```

```python
table1 = pd.DataFrame(tb1)
table2 = pd.DataFrame(tb2)
```

```python
table1
```

```python
table2
```

```python
# left join example :
pd.merge(table1,table2, left_on = 'user_id', right_on = 'u_id', how = 'left')
```

```python
# right join example :
pd.merge(table1,table2, left_on = 'user_id', right_on = 'u_id', how = 'right')
```

# Some Big Problems

```python
customers = pd.read_csv(r"D:\Machine Learning\Joins\sources\customer_table_2.csv")
transactions = pd.read_csv(r"D:\Machine Learning\Joins\sources\transaction_table.csv")
```

```python
customers.head()
```

```python
transactions.head()
```

```python
# Problem Statement - 1

# to get the customers respective region for a respective segment  - calculate the amount summary (sum, avg, mean)

cust = customers[['customer_id','name','region','segment']]
```

```python
# take the need data first for both tables
cust
```

```python
trans = transactions[['customer_id','amount']]
```

```python
trans.head()
```

```python
# now to join the two table with customer_id key

amt_info = pd.merge(trans,cust,left_on='customer_id',right_on = 'customer_id',how = 'left')
```

```python
amt_info.head()
```

```python
# now groupby the cus_id with region, segment and also with calulate the summary 

amt_info.groupby(['customer_id','name','region','segment'])[['customer_id','name','region','segment','amount']].agg(
    {'amount' : 'sum'}
).reset_index()
```

```python
# this is the actual solution

amt_info.groupby(['customer_id','name','region','segment'])[['customer_id','name','region','segment','amount']].agg(
    total_revenue=('amount','sum'),
    avg_revenue=('amount','mean')
).reset_index()
```

```python
# this one is segmenting the region itself for to see the each region segment

amt_info.groupby(['region','segment'])[['customer_id','name','region','segment','amount']].agg(
    total_revenue=('amount','sum'),
    avg_revenue=('amount','mean')
).reset_index()
```

```python
transactions['payment_method'].unique()
```

```python
# Problem Statement - 2

# max revenue of customers in last 6 months via UPI/card 
```

```python
# here changed the date(str) into date time data_type
transactions['date'] = pd.to_datetime(transactions['date'])
```

```python
# now want to calculate the 6 months period of RECORDS

cutoff = transactions['date'].max() - pd.Timedelta(days = 180)
```

```python
cutoff
```

```python
trans_filter = transactions[(transactions['date'] >= cutoff) & (transactions['payment_method'].isin(['UPI','Card']))
                    # Longer Method
                            # (
                            #     (transactions['payment_method'] == 'UPI') |
                            #     (transactions['payment_method'] == 'Card')
                            # )
    ]
```

```python
trans_filter
```

```python
cust_sum = trans_filter.groupby(['customer_id']).agg(
    total_revenue = ('amount','sum'),
    last_purchase = ('date','max')
).reset_index()
```

```python
cust_sum
```

```python
cust_sum['rank'] = cust_sum['total_revenue'].rank(method = 'dense', ascending = False)
```

```python
cust_sum[cust_sum['rank']<=5]
```
