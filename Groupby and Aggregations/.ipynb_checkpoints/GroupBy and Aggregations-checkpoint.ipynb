```python
data.groupby(['customer_name']).agg(
    {'purchase_date':'count'}
).reset_index()
```

```python
data['day'] = pd.to_datetime(data['purchase_date']).dt.day
```

```python
data
```

```python
data.groupby(['customer_name','day']).agg(
    {'purchase_date':'count'}
).reset_index()
```

```python
df = data.sort_values(['customer_name','day'])
```

```python
df
```

```python
df['day_diff'] = df.groupby(['customer_name'])['day'].diff()
```

```python
df
```

```python
df['streak_id'] = (df['day_diff'] != 1).cumsum()
```

```python
df
```

```python
 df['streak_length'] = df.groupby(['customer_name','streak_id'])[['day']].transform('count')
```

```python
df
```

```python
max_streak = (
    df.groupby(['customer_name','streak_id']).size().groupby('customer_name').max().reset_index(name='max_streak')
)
```

```python
max_streak
```

```python
max_streak['rank'] = max_streak['max_streak'].rank(
    method='dense',
    ascending=False
)
```
